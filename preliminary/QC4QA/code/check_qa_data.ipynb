- _improve_answer_span

```python
def _improve_answer_span(doc_tokens, input_start, input_end, tokenizer, orig_answer_text):
    tok_answer_text = " ".join(tokenizer.tokenize(orig_answer_text))
    for new_start in range(input_start, input_end + 1):
        for new_end in range(input_end, new_start - 1, -1):
            text_span = " ".join(doc_tokens[new_start:(new_end + 1)])
            if text_span == tok_answer_text:
                return (new_start, new_end)
    return (input_start, input_end)


# 샘플 문서 토큰
doc_tokens = ["The", "leader", "was", "John", "Smith", "(", "1895", "-", "1943", ")", "."]
# 정답 텍스트 (어노테이션)
orig_answer_text = "1895"
# 원래 어노테이션 범위: (6~9) → "(1895 - 1943)"
input_start = 5   # "("
input_end = 9     # "1943"
new_start, new_end = _improve_answer_span(doc_tokens, input_start, input_end, tokenizer, orig_answer_text)
print("Before:", input_start, input_end, "->", doc_tokens[input_start:input_end+1])
print("After :", new_start, new_end, "->", doc_tokens[new_start:new_end+1])
```

- check_is_max_context

```python
def _check_is_max_context(doc_spans, cur_span_index, position):
    """
    슬라이딩 윈도우로 나눈 문맥 span 중, 
    특정 토큰이 "가장 문맥이 좋은 위치"에 포함되어 있는지 판별
    """
    best_score = None
    best_span_index = None
    for (span_index, doc_span) in enumerate(doc_spans):
        end = doc_span.start + doc_span.length - 1
        if position < doc_span.start:
            continue
        if position > end:
            continue
        num_left_context = position - doc_span.start
        num_right_context = end - position
        print('span_index: ',span_index, 'num_left_context: ',num_left_context, 'num_right_context: ',num_right_context)
        score = min(num_left_context, num_right_context) + 0.01 * doc_span.length
        if best_score is None or score > best_score:
            best_score = score
            best_span_index = span_index
        print('current best_span_index: ', best_span_index, 'best_score: ', best_score)
        print("-"*50)
    return cur_span_index == best_span_index

# DocSpan 구조체 정의
_DocSpan = collections.namedtuple("DocSpan", ["start", "length"])

# 전체 문서 길이 (단어 11개라 가정)
doc_spans = [
    _DocSpan(start=0, length=5),   # Span A → 토큰 인덱스 0~4
    _DocSpan(start=3, length=5),   # Span B → 토큰 인덱스 3~7
    _DocSpan(start=6, length=5)    # Span C → 토큰 인덱스 6~10
]

# cur_span_index: doc_spans 중에서 지금 확인하려는 span의 번호(index)
# 'bought' 토큰 위치 = index 7 = 문서에서 7번째 토큰
token_position = 7
print("Span A:", _check_is_max_context(doc_spans, 0, token_position)) # Span A 검사
print("*"*50)
print("Span B:", _check_is_max_context(doc_spans, 1, token_position)) # Span B 검사
print("*"*50)
print("Span C:", _check_is_max_context(doc_spans, 2, token_position)) # Span C 검사
# Span C가 최적의 문맥 span
```
